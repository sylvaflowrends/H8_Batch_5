Dalam contoh kasus pertama kali ini, Kita akan mencoba membangun sebuah graf dengan struktur sebagai berikut.

Dalam merepresentasikan bentuk graf tersebut, kita akan membuat 2 fungsi untuk memudah

```python
## Blok program ini digunakan untuk mendefinisikan node dan hubungan Antara node dalam bentuk obyek.
graph = { "a" : {"c"},
          "b" : {"c", "e"},
          "c" : {"a", "b", "d", "e"},
          "d" : {"c"},
          "e" : {"c", "b"},
          "f" : {}
        }
```

```python
## Blok program ini digunakan untuk mentransformasikan variabel yang didefinisikan sebelumnya menjadi graph.
def generate_edges(graph):
    edges = []
    for node in graph:
        for neighbour in graph[node]:
            edges.append({node, neighbour})

    return edges
```

```python
# Blok program ini digunakan untuk memanggil fungsi yang telah didefinisikan, dengan menggunakan fungsi print
#beserta nama fungsi yang dibuat.
print(generate_edges(graph))
```

##### penjelasan :
Dapat dilihat dari output diatas, dimana masing-masing informasi mendefiniskan edges dari graph yang dibuat.

```python
## Blok program ini digunakan untuk mendefinisikan sebuah fungsi untuk mengetahui node mana yang tidak memiliki edge,
## dengan nama fungsi "find_isolated_nodes"
def find_isolated_nodes(graph):
    """ returns a set of isolated nodes. """
    isolated = set()
    for node in graph:
        if not graph[node]:
            isolated.add(node)
    return isolated
```

```python
# Blok program ini digunakan untuk memanggil fungsi yang telah didefinisikan, dengan menggunakan fungsi print
#beserta nama fungsi yang dibuat yaitu "find_isolated_nodes".
print(find_isolated_nodes(graph))
```

##### penjelasan :
Dan terlihat pada output diatas, node 'f' tidak memiliki edge sesuai dengan ilustrasi gambar graph diatas.

#### Contoh kasus 2 
Dalam contoh kasus kedua kali ini, Kita akan mencoba membangun sebuah graf dan mendapatkan informasi lebih detail 
mengenai graf tersebut. Untuk memudahkan, Kita akan mendefinisikan sebuah Class dalam bahasa pemrograman python, 
dimana Class ini berisi beberapa fungsi yang bisa kita gunakan untuk mengetahui informasi detail mengenai stuktur dari graf yang kita miliki

```python
""" A Python Class
A simple Python graph class, demonstrating the essential 
facts and functionalities of graphs.
"""
## Blok program ini digunakan untuk mendefinisikan sebuah class bernama yang didalam nya terdapat beberapa fungsi,
## terkait eksploitasi atau transformasi dari graph tersebut.

class Graph(object):

    def __init__(self, graph_dict=None):
        """ initializes a graph object 
            If no dictionary or None is given, 
            an empty dictionary will be used
        """
        if graph_dict == None:
            graph_dict = {}
        self._graph_dict = graph_dict

    def edges(self, vertice):
        """ returns a list of all the edges of a vertice"""
        return self._graph_dict[vertice]
        
    def all_vertices(self):
        """ returns the vertices of a graph as a set """
        return set(self._graph_dict.keys())

    def all_edges(self):
        """ returns the edges of a graph """
        return self.__generate_edges()

    def add_vertex(self, vertex):
        """ If the vertex "vertex" is not in 
            self._graph_dict, a key "vertex" with an empty
            list as a value is added to the dictionary. 
            Otherwise nothing has to be done. 
        """
        if vertex not in self._graph_dict:
            self._graph_dict[vertex] = []

    def add_edge(self, edge):
        """ assumes that edge is of type set, tuple or list; 
            between two vertices can be multiple edges! 
        """
        edge = set(edge)
        vertex1, vertex2 = tuple(edge)
        for x, y in [(vertex1, vertex2), (vertex2, vertex1)]:
            if x in self._graph_dict:
                self._graph_dict[x].add(y)
            else:
                self._graph_dict[x] = [y]

    def __generate_edges(self):
        """ A static method generating the edges of the 
            graph "graph". Edges are represented as sets 
            with one (a loop back to the vertex) or two 
            vertices 
        """
        edges = []
        for vertex in self._graph_dict:
            for neighbour in self._graph_dict[vertex]:
                if {neighbour, vertex} not in edges:
                    edges.append({vertex, neighbour})
        return edges
    
    def __iter__(self):
        self._iter_obj = iter(self._graph_dict)
        return self._iter_obj
    
    def __next__(self):
        """ allows us to iterate over the vertices """
        return next(self._iter_obj)

    def __str__(self):
        res = "vertices: "
        for k in self._graph_dict:
            res += str(k) + " "
        res += "\nedges: "
        for edge in self.__generate_edges():
            res += str(edge) + " "
        return res
```

```python
## Blok program ini digunakan untuk mendefinisikan sebuah variabel yang berisi informasi dari graph yang akan dibuat.
g = { "a" : {"d"},
      "b" : {"c"},
      "c" : {"b", "c", "d", "e"},
      "d" : {"a", "c"},
      "e" : {"c"},
      "f" : {}
    }
```

```python
## Blok program ini digunakan untuk mendefinisikan sebuah variabel yang digunakan untuk mentranformasikan
## variabel tersebut menjadi graph, serta menampilkan dari edge dari graph.
graph = Graph(g)

for vertice in graph:
    print("Edges of vertice {vertice}: ", graph.edges(vertice))
```

```python
## Blok program ini digunakan untuk menambahkan suatu edge, dengan memanggil fungsi '.add_edge'.
graph.add_edge({"ab", "fg"})
graph.add_edge({"xyz", "bla"})
```

```python
## Blok program ini digunakan untuk menampilkan edge dan node yang dimiliki oleh graph tersebut.
print("")
print("Vertices of graph:")
print(graph.all_vertices())

print("Edges of graph:")
print(graph.all_edges())
```

##### penjelasan :
Dapat dilihat dari output vertices of graph, dimana node bertambah 4 yaitu terdapat node "ab", "fg", "xyz", dan "bla".
Serta terdapat pendefinisian dari masing-masing edge nya.

Dalam contoh kasus ketiga kali ini, Kita akan mencoba membangun sebuah graf dengan menggunakan library networkx dan matplotlib

```python
## Blok program ini digunakan untuk mengimport 2 library yaitu library networkx dan matplotlib.
import networkx as nx
import matplotlib.pyplot as plt
```

```python
## Blok program ini digunakan mendefinisikan sebuah variabel untuk membuat graph.
G = nx.Graph() # Right now G is empty
```

```python
## Blok program ini digunakan untuk menambahkan node pada graph G.

G.add_node(1) #add_node digunakan untuk menambahkan node satu-persatu.
G.add_nodes_from([2,3]) #add_node_from digunakan untuk menambahkan lebih dari 1 node.
```

```python
## Blok program ini digunakan untuk menambahkan edge.

G.add_edge(1,2) #add_edge digunakan untuk menambahkan edge satu-persatu.
```

```python
## Blok program ini digunakan untuk menambahkan edge dengan cara yang berbeda.

e = (2,3)
G.add_edge(*e) #digunakan untuk menambahkan edge dengan mengambil nilai dari sebuah variabel bertipe tuple, & * unpacks the tuple.
G.add_edges_from([(1,2), (1,3)]) #add_edge_from digunakan untuk menambahkan lebih dari 1 edge.
```

```python
## Blok program ini digunakan untuk menampilkan node yang dimiliki dari graph yang dibuat.
G.nodes()
```

```python
## Blok program ini digunakan untuk menampilkan edge yang dimiliki dari graph yang dibuat.
G.edges()
```

```python
## Blok program ini digunakan untuk menambahkan beberapa edge baru.

G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(1, 4)
G.add_edge(1, 5)
```

```python
nx.draw(G, with_labels = True) # digunakan untuk memvisualisasikan graph tersebut dengan funsgi 'draw'
```

```python
plt.savefig("contoh-graf-1.png") # digunakan untuk menyimpan hasil gambar graph tersebut
```
