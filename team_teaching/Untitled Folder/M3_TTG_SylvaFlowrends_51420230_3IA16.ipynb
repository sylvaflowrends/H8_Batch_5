```python
#Blok program dibawah ini digunakan untuk mendefinisikan graph diatas dengan menggunakan tipe data dictionary.             
graph = {
    'A' : ['Z', 'S', 'T'],
    'Z' : ['O'],
    'O' : ['S1'],
    'S1' : ['F', 'R'],
    'F' : ['B'],
    'B' : [],
    'R' : [],
    'S' : ['O1', 'F1', 'R1'],
    'O1' : [],
    'F1' : ['B1'],
    'B1' : [],
    'R1' : ['C', 'P'],
    'C' : [],
    'P' : [],
    'T' : ['L'],
    'L' : ['M'],
    'M' : ['D'],
    'D' : []
}
```

```python
#digunakan untuk menyimpan data-data node yang telah dilewati.
visited = set() #Set to keep track of visited nodes of graph
```

```python
#Blok program dibawah ini menjabarkan suatu fungsi bernama dfs
def dfs(visited, graph, node): #function for dfs
    if node not in visited:  #pada line ini didefinisikan if, dimana jika node belum dilewati/visit
        print(node)  #maka akan mencetak node tersebut
        visited.add(node)  #kemudian akan menambahkan data tersebut kedalam variabel visited
        for neighbour in graph[node]:  #setelah itu, akan mengecek node tetangga dari graph tersebut dengan fungsi yg sama.     
            dfs(visited, graph, neighbour)
```

```python
print("Following is the Depth-First Search") #digunakan untuk mencetak kalimat yg berada pada tanda petik
dfs(visited, graph, 'A') # akan menapilkan jalannya dfs, dimana akan mencetak node yang dilewati dari graph tersebut,          
                          # dengan dimulai dari node A.
```
